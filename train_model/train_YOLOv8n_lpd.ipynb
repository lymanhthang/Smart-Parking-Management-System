Tiền xử lý dữ liệu: gộp 2 class thành 1 class duy nhất

```python
import os

label_dir = "/content/drive/MyDrive/Do_An_Nganh/lp_detect/data/test/labels"  # Đường dẫn thư mục chứa file label

for filename in os.listdir(label_dir):
    if filename.endswith(".txt"):
        filepath = os.path.join(label_dir, filename)
        with open(filepath, "r") as f:
            lines = f.readlines()

        new_lines = []
        for line in lines:
            parts = line.strip().split()
            if parts:
                parts[0] = "0"  # Gộp tất cả thành class 0
                new_lines.append(" ".join(parts) + "\n")

        with open(filepath, "w") as f:
            f.writelines(new_lines)
```

Cấu hình file data.yaml

```python
with open('/content/drive/MyDrive/Do_An_Nganh/lp_detect/data/data.yaml', 'w') as f:
    f.write("train: /content/drive/MyDrive/Do_An_Nganh/lp_detect/data/train\n")
    f.write("val: /content/drive/MyDrive/Do_An_Nganh/lp_detect/data/val\n")
    f.write("test: /content/drive/MyDrive/Do_An_Nganh/lp_detect/data/test\n")
    f.write("\n")
    f.write("nc: 1\n")
    f.write("names: ['LicensePlate']\n")
```

Huấn luyện mô hình

```python
from ultralytics import YOLO

model=YOLO('/content/drive/MyDrive/Do_An_Nganh/lp_detect/yolov8n.pt')

model.train(
    data='/content/drive/MyDrive/Do_An_Nganh/lp_detect/data/data.yaml',
    epochs=100,
    imgsz=,
    batch=16,
    name='lp_detector',
    device=0,
    workers=2,
    optimizer='SGD',
    lr0=0.01,
    weight_decay=0.0005,
    momentum=0.937,
    dropout=0.0,
    pretrained=True,
    verbose=True,
    patience=50,
    cos_lr=True,
    close_mosaic=10
)
```
